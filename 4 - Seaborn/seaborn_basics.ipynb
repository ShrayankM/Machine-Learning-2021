```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Scatter Plots

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.info()
```

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'training level', palette = 'viridis');
```

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.scatterplot(x = 'salary', y = 'sales', data = df, size = 'work experience', palette = 'viridis', alpha = 0.3);
```

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.scatterplot(x = 'salary', y = 'sales', data = df, style = 'level of education', hue = 'level of education', palette = 'Dark2');
```

# Distribution Plots

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.rugplot(x = 'salary', data = df);
```

```python
plt.figure(dpi = 125)
sns.displot(x = 'salary', data = df, bins = 20, color = 'orange', edgecolor = 'white', height = 8);
```

```python
plt.figure(figsize = (6, 5), dpi = 125)
sns.histplot(x = 'salary', data = df, bins = 20, kde = True);
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 300)
```

```python
sample_ages = pd.DataFrame(sample_ages, columns = ['age'])
```

```python
plt.figure(figsize = (6, 5), dpi = 125)
sns.histplot(x = 'age', data = sample_ages, bins = 20, edgecolor = 'white', kde = True);
```

```python
plt.figure(figsize = (6, 5), dpi = 125)
sns.kdeplot(x = 'age', data = sample_ages);
```

```python
plt.figure(figsize = (6, 5), dpi = 125)
sns.kdeplot(x = 'age', data = sample_ages, clip = [0, 100], bw_adjust = 0.1, shade = True);
```

# Categorical Plots

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.countplot(x = 'division', data = df);
```

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.countplot(x = 'division', data = df, hue = 'level of education', palette = 'Set2');
```

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.barplot(x = 'level of education', y = 'salary', data = df, estimator = np.max, ci = 'sd', hue = 'division');
plt.legend(bbox_to_anchor = (1, 1));
```

```python
df.groupby('level of education').max()['salary'].sort_values(ascending = False)
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head(6)
```

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.boxplot(x = 'race/ethnicity', y = 'math score', data = df);
```

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.boxplot(y = 'parental level of education', x = 'math score', data = df, hue = 'gender');
plt.legend(bbox_to_anchor = (1, 1));
```

```python
plt.figure(figsize = (10, 4), dpi = 125)
sns.violinplot(x = 'race/ethnicity', y = 'math score', data = df);
```

```python
plt.figure(figsize = (7, 5), dpi = 125)
sns.violinplot(y = 'parental level of education', x = 'reading score', data = df, hue = 'test preparation course', split = True, palette = 'viridis');
plt.legend(bbox_to_anchor = (1, 1));
```

```python
sns.jointplot(x = 'reading score', y = 'writing score', data = df, kind = 'scatter', hue = 'gender', height = 8);
```

```python
sns.pairplot(data = df, height = 4, aspect = 1.2, hue = 'gender', corner = True);
```

```python
#plt.figure(figsize = (7, 5), dpi = 125)
sns.catplot(x = 'gender', y = 'math score', data = df, kind = 'box', col = 'lunch', row = 'test preparation course');
```
