```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 20)
y = 2 * x
```

```python
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title')
plt.xlim(0, 15)
plt.ylim(0, 40)
```

```python
a = np.linspace(start = 0, stop = 10, num = 11, dtype = 'int64')
b = a ** 2
```

```python
fig = plt.figure()

# Outer Axes
outer_axes = fig.add_axes([0, 0, 2, 2])
outer_axes.plot(a, b)

outer_axes.set_xlim(1, 9)
outer_axes.set_ylim(1, 90)
outer_axes.set_xlabel('A')
outer_axes.set_ylabel('B')
outer_axes.set_title('Main Graph')

# Inner Axes
inner_axes = fig.add_axes([0.1, 0.7, 0.8, 1])
inner_axes.plot(a, b)

inner_axes.set_xlim(1, 5)
inner_axes.set_ylim(1, 30)
inner_axes.set_xlabel('A')
inner_axes.set_ylabel('B')
inner_axes.set_title('Zoomed Graph')

plt.show()
```

```python
fig = plt.figure(figsize = (5, 2), dpi = 120)
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(a, b);

fig.savefig('fig_one.png', bbox_inches = 'tight')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

fig.dpi = 125
fig.set_figheight(3)
fig.set_figwidth(8)

axes[0].plot(a, b)
axes[0].set_xlim(1, 10)

axes[1].plot(x, y)
```

```python
x = np.linspace(start = 0, stop = 11, num = 11, dtype = 'int64')
y = x + 2
z = x + 3
```

```python
fig = plt.figure(figsize = (5, 3), dpi = 125)
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x, label = 'x vs x', color = '#00b82b', linewidth = 0.7, linestyle = 'dotted')
l2 = axes.plot(x, y, label = 'x vs y', color = '#db22f0')

l2[0].set_dashes([3, 1, 1, 1, 6, 3])

axes.plot(x, z, label = 'x vs z', color = '#001fb8', marker = 'o', linewidth = 0.7, markersize = 5.6, markeredgecolor = 'white')

axes.legend(loc = [1.1, 1]);
```
